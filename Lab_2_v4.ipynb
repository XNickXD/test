# 1. Графики количества осадков по годам в разных штатах

```python
df_graph = df.groupby(['SUBDIVISION'])
fig,ax = plt.subplots(figsize=(8,6))
for label, df_new in df_graph:
    df_new.plot(kind='line', ax=ax, label=label, x='YEAR', y='ANNUAL', figsize=(20,10))
plt.show()
```

Исходя из полученных графиков можно сделать вывод, что количество осадков, в разных шатах за выборку последних лет достаточно сильно отличаются друг от друга. Полученные графики помогают наглядно визуализировать данную информацию.

# 2. Графики количества осадков по месяцам по всем годам кумулятивно 

```python
df_cum = df[['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']]
df_cum = df_cum.cumsum()
df_cum.plot(figsize=(8,8))
plt.show()
```

Т.к. куммулятивная кривая количества осадков по месяцам данной выборки имеет примерно линейный закон распределения, то ежегодно в рассматриваемых штатах выпадает примерно равное количество осадков.

# 9. Вычислить корреляцию Пирсона и Спирмена между признаками NOV и MAR.

```python
print(df[['NOV','MAR']].corr(method='pearson')) 
print(df[['NOV', 'MAR']].corr(method='spearman'))
sns.jointplot(x='NOV', y='MAR', data=df); 
plt.show();
from scipy.stats import spearmanr as spr
spr(df['NOV'],df['MAR'], nan_policy = 'omit')
```

Вывод: Корреляция - взаимосвязь случайных величин. Чем больше корреляция между признаками, тем больше зависимость между ними. Исходя из вычислений видно, что рассматриваемые признаки (месяцы март и ноябрь) являются некоррелирующими, т.к. коэфициент корреляции по Пирсону = 0,008; а по Спирману = 0,14. Значение p-value мало, это означает что мы отвергаем нулевую гипотезу. Коэффициент корреляция между признаками очень мал, значит можно сделать вывод что корреляция слабо выражена.
